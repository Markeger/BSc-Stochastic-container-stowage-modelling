```julia
using Plots
using JuMP
using StatsPlots
using Statistics

include("scenarios_instance_reader.jl")
include("ship_reader.jl")
include("Fixed_Seeds_Deterministic.jl")
include("DeterministicModel_Simple.jl")
include("ReleaseGenerator.jl")

using .seeds_deterministic 

data_file_ship = "Small_ship.txt"
data_ship = read_ship_instance(data_file_ship)

seed = seeds_deterministic.seed

FactorK_O = 10
```

```julia
i = 2
t = Float64[]

for q in seed
    if i == 11
        i = 2
    end
    data_file = "/Users/konggullerod/Documents/Dokumenter/Skole - DTU/7. Semester/Bachelor projekt/Model/Deterministic_Data/Mixed_Voyage/Small_ship/S_D_2_False_None_18032585.txt"
    data = read_scenario_instance(data_file)

    R = create_release(data, data_ship; release_fraction = 0.80)

    println("Running instance: ", i, " seed: ", q)

    model, x_20, x_40 = build_deterministic_model_simple(data, R, data_ship)
    set_silent(model)
    optimize!(model)
    push!(t, solve_time(model))
    
    println("\nTime: ", t[end])

    i += 1
end
```

```julia
t_2 = Float64[]
t_3 = Float64[]
t_4 = Float64[]
t_5 = Float64[]
t_6 = Float64[]
t_7 = Float64[]
t_8 = Float64[]
t_9 = Float64[]
t_10 = Float64[]

i = 2
for time in t
    if i == 11
        i = 2
    end

    if i == 2
        push!(t_2, time)
    elseif i == 3
        push!(t_3, time)
    elseif i == 4
        push!(t_4, time)
    elseif i == 5
        push!(t_5, time)
    elseif i == 6
        push!(t_6, time)
    elseif i == 7
        push!(t_7, time)
    elseif i == 8
        push!(t_8, time)
    elseif i == 9
        push!(t_9, time)
    elseif i == 10
        push!(t_10, time)
    end

    i += 1
end
```

```julia
# Collect vectors in order (2..10)
vects = [t_2, t_3, t_4, t_5, t_6, t_7, t_8, t_9, t_10]
labels = string.(2:10)
means = [mean(v) for v in vects]

# Determine tighter y-ticks based on data range (11 ticks by default)
allvals = vcat(vects...)
ymin = floor(minimum(allvals))
ymax = ceil(maximum(allvals))

yt = range(ymin, ymax, length=11)

# Boxplots left-to-right by vector number with tighter y-grid
boxplot(vects, labels=labels, legend=false, xlabel="Ports", ylabel="Time", title="Solvetime for 2 to 10 ports", yticks=yt, yminorgrid=true, minorgrid=true)

# Overlay mean values as red diamonds (placed according to their y-value = mean)
scatter!(1:length(vects), means, marker=(:diamond,8), color=:red, label="Mean")
savefig("boxplot_deterministic.png")
```
